```python
import vcf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vcf_file = "/home/marina/Desktop/varian_callers_results/results/gatk_final_vcfs/indel.SNP.recalibrated_99.7.vcf.gz"
vcf_reader = vcf.Reader(filename=vcf_file)

# Initialize empty lists to store attribute values
attributes = {
    "QUAL": [],
    "AF": [],
    "DP": [],
    "GQ": []
}

# Extract attribute values from the VCF file
for record in vcf_reader:
    attributes["QUAL"].append(record.QUAL)
    if record.INFO.get('AF'):
        attributes["AF"].extend(record.INFO['AF'])
    if record.INFO.get('DP'):
        attributes["DP"].append(record.INFO['DP'])
    for call in record.samples:
        if call.data.GQ:
            attributes["GQ"].append(call.data.GQ)

# Create plots for each attribute using seaborn
for attribute, values in attributes.items():
    sns.histplot(values, bins=20)
    plt.xlabel(attribute)
    plt.ylabel('Count')
    plt.title(f'Distribution of {attribute}')
    plt.show()
```

```python
vcf_file = "/home/marina/Desktop/varian_callers_results/results/strlka_final_vcfs/strelka_exome_variants.vcf.gz"
vcf_reader = vcf.Reader(filename=vcf_file)

# Initialize empty lists to store attribute values
attributes = {
    "QUAL": [],
    "AF": [],
    "DP": [],
    "GQ": []
}

# Extract attribute values from the VCF file
for record in vcf_reader:
    attributes["QUAL"].append(record.QUAL)
    if record.INFO.get('AF'):
        attributes["AF"].extend(record.INFO['AF'])
    if record.INFO.get('DP'):
        attributes["DP"].append(record.INFO['DP'])
    for call in record.samples:
        if call.data.GQ:
            attributes["GQ"].append(call.data.GQ)

# Create plots for each attribute using seaborn
for attribute, values in attributes.items():
    sns.histplot(values, bins=20)
    plt.xlabel(attribute)
    plt.ylabel('Count')
    plt.title(f'Distribution of {attribute}')
    plt.show()
```

    QUAL: QUAL stands for Variant Quality Score. It represents the Phred-scaled quality score assigned to the variant call. The QUAL score indicates the confidence in the variant call, with higher values indicating higher confidence.

    AF: AF stands for Allele Frequency. It represents the frequency of the alternate allele in the population. The AF value represents the proportion of individuals carrying the alternate allele at a particular genomic position.

    DP: DP stands for Depth of Coverage. It represents the total depth or number of reads covering a variant position. The DP value indicates the sequencing depth or the number of times a specific position has been sequenced.

    GQ: GQ stands for Genotype Quality. It represents the quality of the called genotype for each sample. The GQ value is assigned to each genotype call and indicates the confidence in the assigned genotype, with higher values indicating higher confidence.
